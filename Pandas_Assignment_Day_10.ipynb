## Applying functions to columns and creating new columns

We need to create another column in d3, which contains  a boolean value for each city to indicate whether it's a union territory or not.
- HINT: Chandigarh, Pondicherry and Delhi are only 3 union territories here.


```python
def is_ut(city):
    union_territories = ['Chandigarh', 'Pondicherry', 'Delhi']
    return city in union_territories

d3['is_ut'] = d3['city'].apply(is_ut)
```

```python
print(d3)
```

The temperatures mentioned in ‘Temperature’ column are mentioned in Celsius, we need another column which contains the same in Fahrenheit.

HINT -
- Define a function c_to_f which takes input temp in celsius and returns a value with temperature in Fahrenheit.
- To check: c_to_f(10) should return 50.


```python
def c_to_f(temp_c):
    return (temp_c * 9/5) + 32
```

```python
print(c_to_f(10))
```

```python
d3['temp_farenhiet'] = d3['temperature'].apply(c_to_f)
print(d3)
```

## Groupby

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
        'Value': [10, 15, 12, 18, 11, 20],
        'Count': [1, 1, 1, 1, 1, 1]}
df_groupby = pd.DataFrame(data)
print(df_groupby)
```

```python
df_groupby_sum = df_groupby.groupby('Category')['Value'].sum()
print(df_groupby_sum)
```

```python
df_groupby_count = df_groupby.groupby('Category')['Value'].count()
print(df_groupby_count)
```

```python
df_groupby_min_max = df_groupby.groupby('Category')['Value'].agg(['min', 'max'])
print(df_groupby_min_max)
```

```python
df_groupby_mean_two_cols = df_groupby.groupby(['Category', 'Count'])['Value'].mean()
print(df_groupby_mean_two_cols)
```

```python
df_groupby_min_max_two_cols = df_groupby.groupby(['Category', 'Count'])['Value'].agg(['min', 'max'])
print(df_groupby_min_max_two_cols)
```
