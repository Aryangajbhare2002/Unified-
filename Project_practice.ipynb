```python
brand_counts = Data['Brand'].value_counts()

plt.figure(figsize=(10,6))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=0,explode=(0.2,0,0,0,0))
plt.title('Distribution of Mobile Brands')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(loc=2)
plt.show()
```

```python
year_counts = Data['Year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.bar(year_counts.index, year_counts.values, width=0.2)
plt.xlabel('Year')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions per Year')
plt.show()
```

```python
import seaborn as sns

# Grouping the data to see units sold by Brand, Model, and Year
sales_summary = Data.groupby(['Year', 'Brand', 'Mobile Model'])['Units Sold'].sum().reset_index()

# Using Seaborn to create a faceted bar plot for better clarity
g = sns.catplot(
    data=sales_summary,
    kind='bar',
    x='Year',
    y='Units Sold',
    hue='Brand',
    col='Brand',
    col_wrap=3,
    height=4,
    aspect=1.2
)
g.set_axis_labels('Year', 'Total Units Sold')
g.set_titles('{col_name}')
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Total Units Sold by Brand and Year')
plt.show()
```

```python
# Calculate total amount for each transaction
Data['Total Amount'] = Data['Units Sold'] * Data['Price Per Unit']

# Group by Brand to get total turnover per brand
brand_turnover = Data.groupby('Brand')['Total Amount'].sum().sort_values(ascending=False)

# Calculate total market turnover
total_market_turnover = brand_turnover.sum()
print(f"Total Mobile Market Turnover: ‡{total_market_turnover:,.2f}")

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    brand_turnover,
    labels=brand_turnover.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=[0.1 if i == 0 else 0 for i in range(len(brand_turnover))],
    shadow=False
)
plt.title('Percentage of Total Turnover by Brand (4-Year Period)')
plt.axis('equal')
plt.legend(title="Brands", loc="best")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using relplot to create individual facets for each Brand
g = sns.relplot(
    data=age_brand_turnover,
    x='Customer Age',
    y='Total Amount',
    hue='Brand',
    col='Brand',
    kind='line',
    col_wrap=2,
    height=5,
    aspect=1.5,
    marker='o'
)

g.fig.suptitle('Individual Brand Turnover Analysis by Customer Age', fontsize=16, y=1.02)
g.set_axis_labels('Customer Age', 'Total Amount Spent (‡)')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'Total Amount' exists from the previous calculation
if 'Total Amount' not in Data.columns:
    Data['Total Amount'] = Data['Units Sold'] * Data['Price Per Unit']

# Aggregate data by Customer Age and Brand
age_brand_turnover = Data.groupby(['Customer Age', 'Brand'])['Total Amount'].sum().reset_index()

plt.figure(figsize=(14, 8))
sns.lineplot(data=age_brand_turnover, x='Customer Age', y='Total Amount', hue='Brand', marker='o')

plt.title('Mobile Purchasing Analysis: Total Amount vs. Customer Age by Brand', fontsize=15)
plt.xlabel('Customer Age', fontsize=12)
plt.ylabel('Total Turnover Amount (‡)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Mobile Brand', loc='upper right')
plt.show()
```

### Marketing Insight Analysis
This chart helps identify peak spending ages for each brand:
- **Trend Identification**: Look for spikes in the line to see which age group contributes most to a brand's revenue.
- **Targeting**: If a specific brand (e.g., Apple) shows high turnover in the 30-40 age range, the marketing team can focus premium ad campaigns on that demographic.

```python
# 1. Average Customer Ratings by Brand
plt.figure(figsize=(10, 6))
avg_ratings = Data.groupby('Brand')['Customer Ratings'].mean().sort_values(ascending=False)
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette='coolwarm')
plt.title('Average Customer Ratings by Brand', fontsize=14)
plt.ylabel('Average Rating (1-5)')
plt.ylim(0, 5)
plt.show()
```

```python
# 2. Payment Method Preference
plt.figure(figsize=(10, 6))
payment_counts = Data['Payment Method'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Payment Methods', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
# 3. Monthly Sales Seasonality (Aggregated over all years)
monthly_sales = Data.groupby('Month')['Units Sold'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Units Sold', marker='o', color='green')
plt.title('Total Units Sold by Month (Seasonality)', fontsize=14)
plt.xticks(range(1, 13))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table for City vs Brand based on Units Sold
city_brand_demand = Data.pivot_table(index='City', columns='Brand', values='Units Sold', aggfunc='sum')

plt.figure(figsize=(12, 10))
sns.heatmap(city_brand_demand, annot=True, fmt='g', cmap='YlGnBu', cbar_kws={'label': 'Total Units Sold'})

plt.title('Heatmap: City-wise Mobile Brand Demand (Units Sold)', fontsize=16)
plt.xlabel('Mobile Brand', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.show()
```
